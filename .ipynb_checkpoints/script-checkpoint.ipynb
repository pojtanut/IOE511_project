```python
import functions
import algorithms 
from optSolver import optSolver
import matplotlib.pyplot as plt

import numpy as np
import time
```

```python
x0 = np.array([1.2,1.2])
print(functions.rosen_func(x0)) 
print(functions.rosen_grad(x0)) 
print(functions.rosen_Hess(x0)) 
```

```python
## Define three class objects required in 'optSolver' and their attributes 

class Problem:
    def __init__(self,name,x0,n,compute_f,compute_g,compute_H):
        self.name = name
        self.x0 = x0
        self.n = n
        self.compute_f = compute_f
        self.compute_g = compute_g
        self.compute_H = compute_H
        
class Method:
    def __init__(self,name,step_type,constant_step_size,c_1=None,tau=None,beta=None,m=None):
        self.name = name
        self.step_type = step_type
        self.constant_step_size = constant_step_size
        if step_type == 'Backtracking':
            self.c_1 = c_1
            self.tau = tau
        if name == 'Newton':
            self.beta = beta
        if name == 'L-BFGS':
            self.m = m
        
class Options:
    def __init__(self,term_tol = 1e-6, max_iterations = 1e2, epsilon=1e-6, verbose=False, check_grad=False, check_Hess=False):
        self.term_tol = term_tol
        self.max_iterations = max_iterations  
        self.epsilon = epsilon
        self.verbose = verbose
        self.check_grad = check_grad
        self.check_Hess = check_Hess
```

```python
## The correct way to instantiate an object in the class is as follow:

# x0 for Rosenbrock
x0 = np.array([[1.2], [1.2]])

# x0 for f2
# x0 = np.array([[1], [1]])

## x0 for f3
# x0 = np.zeros((2, 1))
# x0[0, 0] = 1

problem = Problem('rosen',x0,len(x0),functions.rosen_func,functions.rosen_grad,functions.rosen_Hess)
method = Method('Newton','Backtracking',constant_step_size=1, c_1=1e-4, tau=0.5, beta=1e-6, m=10)
options = Options(1e-6,1e3,1e-6,True,True,True) # if using options = Options(), term_tol and max_iterations will set to be default values
```

```python
## Implement the algorithm on a problem
start_time = time.time()
x,f, _, _ = optSolver(problem,method,options)
print(f"\n run time = {time.time()-start_time}")
```

```python
problem_list = ['Rosenbrock', 'f2', 'f3_2', 'f3_10', 'f3_100', 'f3_1000']
alg_list = ['GradientDescent', 'Newton', 'BFGS', 'L-BFGS_2', 'L-BFGS_5', 'L-BFGS_10']

fig, ax = plt.subplots(2, 6, figsize=(70, 30))


for index, problem_name in enumerate(problem_list):
    f_hist = {key: [] for key in alg_list}
    g_hist = {key: [] for key in alg_list}
    time_hist = {key: None for key in alg_list}
    big_f_list = []
    big_g_list = []
    for alg in alg_list:
        print(f"Running {alg} on {problem_name}")
        # Set problem
        if problem_name == 'Rosenbrock':
            x0 = np.array([[1.2], [1.2]])
            problem = Problem('Rosenbrock',x0,len(x0),functions.rosen_func,functions.rosen_grad,functions.rosen_Hess)      
        elif problem_name == 'f2':
            x0 = np.array([[1], [1]])
            problem = Problem('f2',x0,len(x0),functions.f2_func,functions.f2_grad,functions.f2_Hess)
        elif problem_name == 'f3_2':
            x0 = np.zeros((2, 1))
            x0[0, 0] = 1
            problem = Problem('f3',x0,len(x0),functions.f3_func,functions.f3_grad,functions.f3_Hess)
        elif problem_name == 'f3_10':
            x0 = np.zeros((10, 1))
            x0[0, 0] = 1
            problem = Problem('f3',x0,len(x0),functions.f3_func,functions.f3_grad,functions.f3_Hess)
        elif problem_name == 'f3_100':
            x0 = np.zeros((100, 1))
            x0[0, 0] = 1
            problem = Problem('f3',x0,len(x0),functions.f3_func,functions.f3_grad,functions.f3_Hess)
        elif problem_name == 'f3_1000':
            x0 = np.zeros((1000, 1))
            x0[0, 0] = 1
            problem = Problem('f3',x0,len(x0),functions.f3_func,functions.f3_grad,functions.f3_Hess)

        # Set options
        options = Options(1e-6,1e3,True)

        # Set method
        if alg == 'GradientDescent':
            method = Method(alg,'Backtracking',constant_step_size=1, c_1=1e-4, tau=0.5)
        elif alg == 'Newton':
            method = Method(alg,'Backtracking',constant_step_size=1, c_1=1e-4, tau=0.5, beta=1e-6)
        elif alg == 'BFGS':
            method = Method(alg,'Backtracking',constant_step_size=1, c_1=1e-4, tau=0.5, beta=1e-6)
        elif alg == 'L-BFGS_2':
            method = Method('L-BFGS','Backtracking',constant_step_size=1, c_1=1e-4, tau=0.5, beta=1e-6, m=2)
        elif alg == 'L-BFGS_5':
            method = Method('L-BFGS','Backtracking',constant_step_size=1, c_1=1e-4, tau=0.5, beta=1e-6, m=5)
        elif alg == 'L-BFGS_10':
            method = Method('L-BFGS','Backtracking',constant_step_size=1, c_1=1e-4, tau=0.5, beta=1e-6, m=10)

        # Run optim
        start_time = time.time()
        x, f, tmp_f_hist, tmp_g_hist = optSolver(problem,method,options)
        print(f"\nRun time = {time.time()-start_time}")
        print('-'*20)

        f_hist[alg] = tmp_f_hist
        g_hist[alg] = tmp_g_hist

    for k, v in f_hist.items():
        ax[0, index].plot(v, label=k)
        big_f_list += v

    for k, v in g_hist.items():
        ax[1, index].plot(v, label=k)
        big_g_list += v

    ax[0, index].legend()
    ax[1, index].legend()

    ax[0, index].grid(visible=True, alpha=0.2)
    ax[1, index].grid(visible=True, alpha=0.2)

    ax[0, index].set_title(f'Function value across iterations for {problem_name}')
    ax[1, index].set_title(f'Gradient across iterations for {problem_name}')

    ax[0, index].set_xlabel('Iterations')
    ax[0, index].set_ylabel('Function value')

    ax[1, index].set_xlabel('Iterations')
    ax[1, index].set_ylabel('Gradient')

    if not alg.startswith('f3'):
        ax[0, index].set_yscale('log')
        ax[1, index].set_yscale('log')

    ax[0, index].set_ylim(np.min(big_f_list)-1, np.max(big_f_list)+1)
    ax[1, index].set_ylim(np.min(big_g_list)-1, np.max(big_g_list)+1)
plt.savefig('hw2_plot.svg')
```

```python
grad_x = functions.rosen_grad(x)
```

```python
epsilon = 1e-6
grad_approx = np.zeros((2, 1))
for i in range(2):
    e = np.zeros((2, 1))
    e[i] = epsilon

    x_ub = x + e
    x_lb = x - e

    grad_approx[i] = (functions.rosen_func(x_ub) - functions.rosen_func(x_lb)) / (2 * epsilon)
```

```python
np.linalg.norm(grad_x - grad_approx, ord=2)
```
